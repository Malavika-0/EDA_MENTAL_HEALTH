```python
def severity(row):
    if pd.isna(row['Severity']):  # Check if the value is missing (NaN)
        if row['Stress_Level'] == 'High':
            return 'High'
        elif row['Stress_Level'] == 'Medium' and row['Physical_Activity_Hours'] < 2:
            return 'Medium'
        elif row['Consultation_History'] == 'Yes':
            return 'Medium'
        else:
            return 'Low'
    else:
        return row['Severity']  # Retain existing values
df['Severity'] = df.apply(func=severity,axis=1)
```
